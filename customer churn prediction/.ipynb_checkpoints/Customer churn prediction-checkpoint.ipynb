# Customer churn prediction

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/Churn_Modelling.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.drop(columns = ["RowNumber", "Surname"], axis = 1, inplace = True)
df.head()
```

```python
df["Exited"].value_counts()
```

```python
plt.bar(["Active", "Churned"] , df["Exited"].value_counts().values)
plt.title("Customer Churn Analysis. Active vs. Churned customer counts.")
plt.xlabel("Types of customers.")
plt.ylabel("Number of cutomers.")
plt.show()
```

```python
import seaborn as sns
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
```

```python
df.Geography.value_counts()
```

```python
# Create the pie chart
plt.pie(df["Exited"].value_counts().values, labels=["Active customers", "Churned customers"], autopct='%1.1f%%', startangle=30)
plt.title('Percentage of active vs. churned customers')
plt.axis('equal')
plt.show()
```

By looking at the pie chart we can clearly see that about 20% of the customers have stopped using the bank.

```python
plt.pie(df.Geography.value_counts().values, labels=df.Geography.unique(), autopct='%1.1f%%', startangle=30)
plt.title('Percentage of customers from each country')
plt.axis('equal')
plt.show()
```

From the above chart we found out that the all customers belongs to Germany, Spain and France. Perhaps it is a European bank. Most users are from France.

```python
plt.pie(df[df["Exited"] == 1].Geography.value_counts().values, labels=df.Geography.unique(), autopct='%1.1f%%', startangle=30)
plt.title('Percentage of customers churned from each country')
plt.axis('equal')
plt.show()
```

From the above pie chart we can see that 40% of the churned users belongs to France. Because France have more accounts

```python
germany = df[(df["Geography"] == "Germany")]["Exited"].value_counts().values
plt.pie(germany, labels=["Active customers", "Churned customers"], autopct='%1.1f%%', startangle=30)
plt.title('Percentage of active vs. churned customers Germany.')
plt.axis('equal')
plt.show()
```

```python
spain = df[(df["Geography"] == "Spain")]["Exited"].value_counts().values
plt.pie(spain, labels=["Active customers", "Churned customers"], autopct='%1.1f%%', startangle=30)
plt.title('Percentage of active vs. churned customers Spain.')
plt.axis('equal')
plt.show()
```

```python
france = df[(df["Geography"] == "France")]["Exited"].value_counts().values
plt.pie(france, labels=["Active customers", "Churned customers"], autopct='%1.1f%%', startangle=30)
plt.title('Percentage of active vs. churned customers France.')
plt.axis('equal')
plt.show()
```

German account holders have a highest percentage of churning according to above pie charts

```python
active = [2795, 1696, 1850]
churned = [795, 196, 150]

plt.bar(["France", "Spain", "Germany"], active)
plt.bar(["France", "Spain", "Germany"], churned, bottom=active)
plt.legend(["active", "churned"])
plt.show()
```
